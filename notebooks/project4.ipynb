As it shows, we may be able to use feature engineering to select features for prediction.

```python
from sklearn.preprocessing import LabelEncoder

e = LabelEncoder()
e.fit(spi_matches['team1'])
spi_matches['team1'] = e.transform(spi_matches['team1'])
spi_matches['team2'] = e.transform(spi_matches['team2'])

e.fit(spi_matches['date'])
spi_matches['date'] = e.transform(spi_matches['date'])

```

```python
spi_matches= spi_matches.drop('league', 1)
spi_matches= spi_matches.drop('probtie', 1)
spi_matches= spi_matches.dropna()
```
